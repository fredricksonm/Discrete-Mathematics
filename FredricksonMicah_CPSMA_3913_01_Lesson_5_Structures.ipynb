## Structures

Importing library

```python
import graphviz
```

Defining nodes and creating a graph that describes my family's relationship.

```python
dot = graphviz.Digraph(comment='My Family')
dot.node('MF', 'Micah')
dot.node('CF', 'Charissa')
dot.node('JF', 'Jonathan')
dot.node('BF','Brittney')
dot.node('SF','Sally')
dot.node('KF','Ken')
```

Defining edges

```python
dot.edge( 'MF','SF', 'Mom')
dot.edge('MF','KF','Dad')
dot.edge('MF','CF','Sister')
dot.edge('MF','JF','Brother')
dot.edge('MF','BF','Sister')
dot.edge('KF','SF','Wife')
dot.edge('SF','KF','Husband')
dot.edge( 'CF','SF', 'Mom')
dot.edge('CF','KF','Dad')
dot.edge( 'JF','SF', 'Mom')
dot.edge('JF','KF','Dad')
dot.edge( 'BF','SF', 'Mom')
dot.edge('BF','KF','Dad')
dot.edge('KF','MF','Son')
dot.edge('SF','MF','Son')
```

showing graph

```python
dot
```

There is a cycle in the graph since Ken and Sally both point back to Micah.
